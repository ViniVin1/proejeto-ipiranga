```python
import pandas as pd
```

```python
df = pd.read_csv('respostas.csv', header = 0, sep  =',')
display(df)
```

3 – Relação tipo de combustíveis mais consumidos por estado:

      5 – Qual dos combustíveis você mais utiliza?
      3 – De qual estado você é? 

```python
resultado_eletrico = df.query('combustivel == "Elétrico"')[['combustivel','estado']].value_counts()
resultado_gasolinaaditivada = df.query('combustivel == "Gasolina Aditivada"')[['combustivel','estado']].value_counts()
resultado_gnv = df.query('combustivel == "GNV"')[['combustivel','estado']].value_counts()
resultado_diesel = df.query('combustivel == "Diesel"')[['combustivel','estado']].value_counts()
resultado_etanol = df.query('combustivel == "Etanol"')[['combustivel','estado']].value_counts()
resultado_gasolinacomum = df.query('combustivel == "Gasoluna comum"')[['combustivel','estado']].value_counts()
print(resultado_eletrico)
print('-'*100)
print(resultado_gasolinaaditivada)
print('-'*100)
print(resultado_gnv)
print('-'*100)
print(resultado_diesel)
print('-'*100)
print(resultado_etanol)
print('-'*100)
print(resultado_gasolinacomum)
```
